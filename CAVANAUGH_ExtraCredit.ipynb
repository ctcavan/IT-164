```python
import networkx as nx

#create graph with networks
HoD=nx.Graph()

# add characters
HoD.add_nodes_from(['Viserys', 'Rhaenyra', 'Daemon', 'Alicent', 'Rhaenys', 'Larys','Otto', 'Hobert', 'Craghas','Ser Criston Cole', 'Joffrey', 'Aemma', 'Laenor', 'Rhae', 'Laena', 'Corlys', 'Baela', 'Vaemond', 'Harwin', 'Aegon', 'Vaemond', 'Jacaerys', 'Lcuerys', 'Lyonel', 'Ryam', 'Lord Beesbury', 'Harrold', 'Erryk', 'Arryk', 'Lord Beesbury', 'Rhaena', 'Aemond'])
```

```python
#recreate a network for who served whom

Serve=nx.DiGraph()
Serve.add_edges_from([('Otto', 'Viserys'), ('Lyonel', 'Viserys'), ('Larys', 'Alicent')], label = 'served')
ServeLabels=nx.get_edge_attributes(Serve, 'label')
```

```python
#create a network for siblings
Siblings=nx.DiGraph()
Siblings.add_edges_from([('Otto', 'Hobert',), ('Baela', 'Rhaena'), ('Corlys', 'Vaemond'), ('Larys', 'Harwin'), ('Aegon', 'Aemond')], label='siblings')
SiblingsLabels=nx.get_edge_attributes(Siblings,'label')
```

```python
#create a network for who killed who
Killed=nx.DiGraph()
Killed.add_edges_from([('Daemon', 'Craghas'), ('Daemon', 'Vaemond'), ('Daemon', 'Rhae'), ('Ser Criston Cole', 'Lord Beesbury')], label='killed')
KilledLabels=nx.get_edge_attributes(Killed, 'label')
```

```python
#create a network for marriage
Marriage=nx.DiGraph()
Marriage.add_edges_from([('Viserys', 'Aemma'), ('Rhaenyra', 'Laenor'), ('Daemon', 'Rhae'), ('Laena', 'Daemon'), ('Corlys', 'Rhaenys')], label='Married')
MarriageLabels=nx.get_edge_attributes(Marriage, 'label')
```

```python
#create a network for parents
Parents=nx.DiGraph()
Parents.add_edges_from([('Otto', 'Alicent'), ('Rhaenyra', 'Jacaerys'), ('Rhaenyra', 'Lcuerys'), ('Rhaenyra', 'Joffrey'),], label='Parent')
ParentsLabels=nx.get_edge_attributes(Parents, 'label')
```

```python
#create a network for allies
Allies=nx.DiGraph()
Allies.add_edges_from([('Viserys', 'Vaemond'), ('Viserys', 'Corlys')], label='Allies')
AlliesLabels=nx.get_edge_attributes(Allies, 'label')
```

```python
#create a network for guardians
Guardians=nx.DiGraph()
Guardians.add_edges_from([('Ser Criston Cole', 'Alicent'), ('Ryam', 'Viserys'), ('Harrold', 'Viserys'), ('Erryk', 'Viserys'), ('Arryk', 'Viserys')], label='Guardian')
GuardiansLabels=nx.get_edge_attributes(Guardians, 'label')
```

```python
import matplotlib.pyplot as plt
plt.figure(1,figsize=(16,12))
pos=nx.circular_layout(HoD)
nx.draw(HoD, pos, node_color='grey', node_size=5, with_labels=True)

#Draw the serving edges
nx.draw_networkx_edges(Serve, pos, edge_color='purple')
#adding purple serve labels to the graph
nx.draw_networkx_edge_labels(Serve, pos, edge_labels=ServeLabels, label_pos = 0.7,
                             font_color='purple')
#draw sibling edges
nx.draw_networkx_edges(Siblings, pos, edge_color='blue')
#add blue sibling labels to the graph
nx.draw_networkx_edge_labels(Siblings,pos, edge_labels=SiblingsLabels, label_pos = 0.5,font_color='blue')
                            
#Draw the killed edges
nx.draw_networkx_edges(Killed, pos, edge_color='red')
#adding red killed labels to the graph
nx.draw_networkx_edge_labels(Killed,pos,edge_labels=KilledLabels,label_pos = 0.2,font_color='red')

#draw marriage edges
nx.draw_networkx_edges(Marriage, pos, edge_color='green')
#adding green marriage labels to the graph
nx.draw_networkx_edge_labels(Marriage, pos, edge_labels=MarriageLabels, label_pos = 0.2, font_color='green') 

#draw parent edges
nx.draw_networkx_edges(Parents, pos, edge_color='deeppink')
#adding orange parents labels to the graph
nx.draw_networkx_edge_labels(Parents, pos, edge_labels=ParentsLabels, label_pos = 0.2, font_color='deeppink')

#draw ally edges
nx.draw_networkx_edges(Allies, pos, edge_color='brown')
#adding pink ally labels to the graph
nx.draw_networkx_edge_labels(Allies, pos, edge_labels=AlliesLabels, label_pos = 0.5, font_color='brown')

#draw guardian edges
nx.draw_networkx_edges(Guardians, pos, edge_color='gold')
#adding gold guardian labels to the graph
nx.draw_networkx_edge_labels(Guardians, pos, edge_labels=GuardiansLabels, label_pos= 0.5, font_color='gold')



plt.show()
```

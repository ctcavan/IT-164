```python
import networkx as nx

#create graph with networks
HoD=nx.Graph()

# add characters
HoD.add_nodes_from(['Viserys', 'Rhaenyra', 'Daemon', 'Alicent', 'Rhaenys', 'Larys','Otto', 'Hobert'])
```

```python
#recreate a network for who served whom

Serve=nx.DiGraph()
Serve.add_edges_from([('Otto', 'Viserys'), ('Larys', 'Alicent')], label = 'served')
ServeLabels=nx.get_edge_attributes(Serve, 'label')
```

```python
#create a network for siblings
Siblings=nx.DiGraph()
Siblings.add_edges_from([('Otto', 'Hobert'), ('Viserys', 'Daemon')], label='siblings')
SiblingsLabels=nx.get_edge_attributes(Siblings,'label')
```

```python
#create a network for who killed who
Killed=nx.DiGraph()
Killed.add_edges_from([('Viserys', 'Craghas Drahar'), ('Ser Criston Cole', 'Joffrey Lonmouth')], label='killed')
KilledLabels=nx.get_edge_attributes(Killed, 'label')
```

```python
import matplotlib.pyplot as plt
plt.figure(1,figsize=(6,6))
pos=nx.circular_layout(HoD)
nx.draw(HoD, pos, node_color='grey', node_size=5, with_labels=True)

#Draw the serving edges
nx.draw_networkx_edges(Serve, pos, edge_color='purple')
#adding purple serve labels to the graph
nx.draw_networkx_edge_labels(Serve, pos, edge_labels=ServeLabels, label_pos = 0.2,
                             font_color='purple')
#draw sibling edges
nx.draw_networkx_edges(Siblings, pos, edge_color='blue')
#add blue sibling labels to the graph
nx.draw_networkx_edge_labels(Siblings,pos, edge_labels=SiblingsLabels, label_pos = 0.2,font_color='blue')
                            
#Draw the killed edges
nx.draw_networkx_edge_labels(Killed, pos, edge_labels='red')
#adding red killed labels to the graph
n

plt.show()
```
